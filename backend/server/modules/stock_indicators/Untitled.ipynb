```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_ta as pta
# Load stock price data into a Pandas DataFrame (assuming CSV format)
df = pd.read_excel('^NSEI_1d.xlsx')  # Replace 'stock_prices.csv' with your file path

# Assuming your DataFrame has a 'Date' column and a 'Close' column for closing prices
# If not, adjust accordingly
df['date'] = pd.to_datetime(df['date'])  # Convert 'Date' column to datetime
df.set_index('date', inplace=True)  # Set 'Date' column as index

# Define a function to compute the moving average of stock prices
def moving_average(data, window_size):
    """Compute the moving average of data."""
    return data.rolling(window=window_size).mean()

# Smooth the closing prices using a moving average with a window size of 10 days
window_size = 5
df['close_moving'] = pta.sma(df['close'], length = 80, offset=0, presma = True)
df['Close_smoothed'] =pta.sma(df['close_moving'], length = 5, offset=0, presma = True)

ddf = df.iloc[-100:]
# Plot original and smoothed closing prices
plt.figure(figsize=(10, 6))
plt.plot(ddf.index, ddf['close'], label='Original Closing Prices')
plt.plot(ddf.index, ddf['Close_smoothed'], color='red', label='Smoothed Closing Prices (SMA)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Smoothing Stock Prices with Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```
